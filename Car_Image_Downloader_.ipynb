Imports csv with all car makes and models from EPA database.


```
import csv

cardb = []

with open('/content/gdrive/My Drive/car_db.csv', newline='') as csvfile:
  carinfo = csv.reader(csvfile, delimiter=' ', quotechar='|')
  for row in carinfo:
   cardb.append(' '.join(row))
```

Designates Car Images as the path for the next for loop we are about to run.

```
path = Path('/content/gdrive/My Drive/CarPictures')

from bing_image_downloader import downloader
```

Creates a new folder for each individual car model to store the car images to be taken from Bing.

```
for o in cardb: 
    dest = (path/o)
    dest.mkdir(exist_ok=True)
    downloader.download(o, limit=1,  output_dir= path , adult_filter_off=True, force_replace=False, timeout=60)
```
